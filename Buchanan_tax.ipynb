```python
import numpy as np
import matplotlib.pyplot as plt

# Create the figure and axes
fig, ax = plt.subplots(figsize=(10, 6))

# Define demand functions for two goods X1 and X2
def demand_x1(p1):
    return 100 - 2 * p1  # Linear demand for X1

def demand_x2(p2):
    return 80 - 1.5 * p2  # Linear demand for X2

# Generate price and quantity data for plotting
p = np.linspace(0, 50, 100)
x1 = demand_x1(p)
x2 = demand_x2(p)

# Plot demand curves for X1 and X2
ax.plot(p, x1, label='$D_{X1}$: Demanda de $X_1$', color='blue')
ax.plot(p, x2, label='$D_{X2}$: Demanda de $X_2$', color='green')

# Iso-revenue lines (for simplicity, assume linear iso-revenue)
ax.plot(p, 50 - 0.5 * p, linestyle='--', color='red', label='Iso-revenue line 1')
ax.plot(p, 60 - 0.6 * p, linestyle='--', color='orange', label='Iso-revenue line 2')

# Shaded welfare loss areas
p_shade = np.linspace(10, 20, 100)
ax.fill_between(p_shade, demand_x1(p_shade), color='blue', alpha=0.2, label='Pérdida de bienestar $X_1$')
ax.fill_between(p_shade, demand_x2(p_shade), color='green', alpha=0.2, label='Pérdida de bienestar $X_2$')

# Labels and title
ax.set_xlabel('Precio ($P$)', fontsize=12)
ax.set_ylabel('Cantidad ($Q$)', fontsize=12)
ax.set_title('Figura del Apéndice: Tributación, Ingresos y Pérdida de Bienestar', fontsize=14)

# Legend
ax.legend()

# Grid for better visualization
ax.grid(True, linestyle='--', alpha=0.6)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create the figure and axes
fig, ax = plt.subplots(figsize=(10, 6))

# Define demand functions for two goods X1 and X2
def demand_x1(q1):
    return 50 - 0.5 * q1  # Linear demand for X1

def demand_x2(q2):
    return 60 - 0.6 * q2  # Linear demand for X2

# Generate quantity and price data for plotting
q = np.linspace(0, 100, 100)
p1 = demand_x1(q)
p2 = demand_x2(q)

# Plot demand curves for X1 and X2
ax.plot(q, p1, label='$D_{X1}$: Demanda de $X_1$', color='blue')
ax.plot(q, p2, label='$D_{X2}$: Demanda de $X_2$', color='green')

# Iso-revenue lines (for simplicity, assume linear iso-revenue)
ax.plot(q, 25 - 0.25 * q, linestyle='--', color='red', label='Iso-revenue line 1')
ax.plot(q, 30 - 0.3 * q, linestyle='--', color='orange', label='Iso-revenue line 2')

# Shaded welfare loss areas
q_shade = np.linspace(20, 40, 100)
ax.fill_between(q_shade, demand_x1(q_shade), color='blue', alpha=0.2, label='Pérdida de bienestar $X_1$')
ax.fill_between(q_shade, demand_x2(q_shade), color='green', alpha=0.2, label='Pérdida de bienestar $X_2$')

# Labels and title
ax.set_xlabel('Cantidad ($Q$)', fontsize=12)
ax.set_ylabel('Precio ($P$)', fontsize=12)
ax.set_title('Figura del Apéndice: Tributación, Ingresos y Pérdida de Bienestar', fontsize=14)

# Legend
ax.legend()

# Grid for better visualization
ax.grid(True, linestyle='--', alpha=0.6)

# Show the plot
plt.tight_layout()
plt.show()
```
